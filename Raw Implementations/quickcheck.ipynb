```python
def word_to_byte_symbols(word: str) -> tuple:
    word_bytes = tuple(bytes([b]) for b in word.encode("utf-8"))
    marker = (b"</w>",)
    return word_bytes + marker
```

```python
mb=word_to_byte_symbols("café")
print(mb)
```
